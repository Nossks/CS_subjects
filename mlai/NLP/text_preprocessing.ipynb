## Stemming

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()

from nltk.stem import RegexpStemmer
rs=RegexpStemmer('ing$|s$|e$|able$', min=4)

from nltk.stem import SnowballStemmer
ss=SnowballStemmer("english")
```

```python
words=["eating","eats","eaten","play","plays","playing","history","congratulations"]
```

```python
for word in words:
    print(word,"------>",ps.stem(word))
```

```python
for word in words:
    print(word,"------>",rs.stem(word))
```

```python
for word in words:
    print(word,"------>",ss.stem(word))
```

## lemmatization

```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
```

```python
for word in words:
    print(word,"------>",wnl.lemmatize(word,pos="v"))
```

## final example

```python
corpus="""Good morning, everyone!
Today, I want to remind you of something important: you are capable of achieving greatness. Life is full of challenges, and sometimes it feels like the odds are stacked against us. But the truth is, every challenge you face is an opportunity to grow, to learn, and to become stronger than you were yesterday.
Remember, success is not about never failing; it’s about rising every time you fall. Each setback is a stepping stone on the path to your dreams. Embrace your mistakes, for they are proof that you are trying, that you are daring to chase your goals.
Never let fear hold you back. The road to success isn’t easy, but it’s worth it. Believe in yourself, even when the world seems doubtful. Take small steps every day, and over time, they will lead to giant leaps.
Most importantly, don't forget to celebrate your progress, no matter how small. You are your greatest supporter. So stand tall, face your challenges with courage, and never stop striving for the best version of yourself.
The future is yours to shape—go out there and make it extraordinary."""
```

```python
print(corpus)
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()

sentences=sent_tokenize(corpus)

stp_wrd=set(stopwords.words("english"))
```

```python
for i in sentences:
    print([wnl.lemmatize(word) for word in word_tokenize(i) if word not in stp_wrd])
```
