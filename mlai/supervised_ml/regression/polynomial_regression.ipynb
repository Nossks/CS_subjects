```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt;
```

```python
X=6*np.random.rand(100,1)-3
y=0.5*X**2+1.5*X+2+np.random.randn(100,1)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split;
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures;
poly=PolynomialFeatures(degree=3,include_bias=True)
X_train_poly=poly.fit_transform(X_train);
X_test_poly=poly.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression;
regression=LinearRegression(n_jobs=-1)
```

```python
modal=regression.fit(X_train_poly,y_train)
```

```python
from sklearn.metrics import r2_score;
y_pred=modal.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
plt.scatter(X_train,modal.predict(X_train_poly))
plt.scatter(X_train,y_train)
```

```python
from sklearn.pipeline import Pipeline
```

PIPELINE

```python
def polynomial_regression(degree):
    X_new=np.linspace(-3,3,200).reshape(200,1)

    poly_features=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression(n_jobs=-1)
    polynomial_regression=Pipeline([
        ("features decided",poly_features),
        ("model selection",lin_reg)
    ])
    polynomial_regression.fit(X_train,y_train)
    y_pred_new=polynomial_regression.predict(X_new)
    #ploting
    plt.plot(X_new,y_pred_new,'r',label="Degree"+str(degree),linewidth=3)
    plt.plot(X_train,y_train,'b.',linewidth=3)
    plt.plot(X_test,y_test,'g.',linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis([-4,4,0,10])
    plt.show()
```

```python
polynomial_regression(5)
```
