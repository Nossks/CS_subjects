```python
import seaborn as sns
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.columns
```

```python
X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
X.head()
```

```python
y=df[["total_bill"]]
y.head()
```

```python
df["time"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=3)
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.preprocessing import LabelEncoder

le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()

X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])

X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

ohe1 = OneHotEncoder(sparse_output=False)
encoded_data = ohe1.fit_transform(X_train[["day"]])
df_encoded = pd.DataFrame(encoded_data, columns=ohe1.get_feature_names_out(["day"]))

df_encoded.reset_index(drop=True, inplace=True)
X_train.reset_index(drop=True, inplace=True)

X_train.drop(labels="day", axis=1, inplace=True)
X_train = pd.concat([df_encoded, X_train], axis=1)
```

```python
X_train.head()
```

```python
encoded_data = ohe1.transform(X_test[["day"]]) 
df_encoded = pd.DataFrame(encoded_data, columns=ohe1.get_feature_names_out(["day"]))

df_encoded.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)

X_test.drop(labels="day", axis=1, inplace=True)
X_test = pd.concat([df_encoded, X_test], axis=1)
```

```python
X_test.head()
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
svr=SVR()
para={
    "kernel":['rbf', 'sigmoid'],
    "C":[1,10,100,1000],
    "gamma":[.0001,.1,.01,.001]
}
grid=GridSearchCV(estimator=svr,param_grid=para,verbose=3,refit=True)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
