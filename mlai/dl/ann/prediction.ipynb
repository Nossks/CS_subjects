```python
from tensorflow.keras.models import load_model
import pickle
import tensorflow as tf
import pandas as pd
```

```python
model=load_model("model.h5")

with open('one_hot_encoder.pkl',"rb") as f:
  ohe=pickle.load(f)

with open('standard_scalar.pkl',"rb") as f:
  ss=pickle.load(f)

with open('label_encoder.pkl',"rb") as f:
  lbl=pickle.load(f)
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
loss = tf.keras.losses.BinaryCrossentropy()
model.compile(optimizer=opt, loss=loss, metrics=["accuracy"])
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
input_data["Gender"]=lbl.transform([input_data["Gender"]])[0]
df=pd.DataFrame([input_data])
arr=ohe.transform([[input_data["Geography"]]])
encoded_geo=pd.DataFrame(arr,columns=ohe.get_feature_names_out())
df=pd.concat([df,encoded_geo],axis=1)
df.drop(columns=["Geography"],inplace=True)
df

X=ss.transform(df)
X
```

```python
model.predict(X)   ##without compilong
```
