# Scalar

```python
import torch

x = torch.tensor(3)
y = torch.tensor(9)

w = torch.rand(1,requires_grad=True)
b = torch.rand(1,requires_grad=True)

z = torch.sigmoid(w*x+b)
y_pred = z
loss = torch.abs(y_pred-y)
```

```python
print(w.grad,b.grad)
loss.backward()
print(w.grad,b.grad)
```

# vector

```python
x = torch.ones(3,requires_grad=True)
y = torch.relu((x**3).mean())

print(x.grad)
y.backward()
print(x.grad)
```

# clearing gradient

```python
x = torch.tensor([2],requires_grad=True,dtype=torch.float16)

y = x**2

y.backward()
print(x.grad)
x.grad.zero_
```

```python
print(x.grad)
```
