# GPU tensors

```python
import time
```

```python
import time
import torch

n = 10000
x = torch.rand(n,n)
y = torch.rand(n,n)

start_cpu_time = time.time()
z = torch.matmul(x,y)
total_cpu_time = time.time()-start_cpu_time

device = torch.device('cuda')
a = x.to(device)
b = torch.rand(n,n,device=device)

start_gpu_time = time.time()
c = torch.matmul(a,b)
torch.cuda.synchronize()
total_gpu_time = time.time()-start_gpu_time

print(f"diff is {total_cpu_time-total_gpu_time}")
print(f"for cpu time is {total_cpu_time}")
print(f"for gpu time is {total_gpu_time}")
```
