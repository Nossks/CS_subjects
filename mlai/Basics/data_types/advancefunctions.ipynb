```python
def call_back_func(call,lst):
    return [call(i) for i in lst]

print(call_back_func(lambda num:num>5 , [1,2,3,4,5,6]))
```

```python
def fun2(lst):
    return map(lambda num:num*num,lst);

print(list(fun2([1,2,3,4,5])))
```

```python
def high(fill,mp,lst):
    return [mp(num) for num in lst if fill(num)];

print(high(lambda num:num>=5,lambda num:num*num,[1, 2, 3, 4, 5,6,3,7]))
```
