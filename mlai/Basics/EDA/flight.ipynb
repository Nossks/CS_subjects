```python
import pandas as pd;
import numpy as np;
from sklearn.preprocessing import OneHotEncoder;
encoder=OneHotEncoder();
```

```python
df=pd.read_excel("flight_price.xlsx");
df.info()
```

```python
df.head(2)
```

```python
df["date"]=df["Date_of_Journey"].str.split("/").str[0];
df["date"]=df["date"].astype(int);
df["month"]=df["Date_of_Journey"].str.split("/").str[1];
df["month"]=df["date"].astype(int);
df["year"]=df["Date_of_Journey"].str.split("/").str[2];
df["year"]=df["date"].astype(int);
```

```python
df.drop("Date_of_Journey",axis=1,inplace=True)
df.drop("Route",axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df["Dep_Hour"]=df["Dep_Time"].str.split(":").str[0]
df["Dep_Min"]=df["Dep_Time"].str.split(":").str[1]
```

```python
df.drop("Dep_Time",axis=1,inplace=True)
```

```python
df["Dep_Hour"]=df["Dep_Hour"].astype(int)
df["Dep_Min"]=df["Dep_Min"].astype(int)
df.info()
```

```python
df.head()
```

```python
df["Arrival_Time"]=df["Arrival_Time"].str.split(" ").str[0]
```

```python
df["Arrival_Hour"]=df["Arrival_Time"].str.split(":").str[0]
df["Arrival_Min"]=df["Arrival_Time"].str.split(":").str[1]
```

```python
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df["Arrival_Hour"]=df["Arrival_Hour"].astype(int)
df["Arrival_Min"]=df["Arrival_Min"].astype(int)
df.info()
```

```python
df.head()
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].mode()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,np.nan:1, '4 stops':4})
```

```python
df["Total_Stops"].unique()
df.head(2)
```

```python
df["Duration_Hours"] = df["Duration"].str.extract(r"(\d+)h", expand=False).fillna(0).astype(int)
df["Duration_Min"] = df["Duration"].str.extract(r"(\d+)m", expand=False).fillna(0).astype(int)
```

```python
df.head(1)
```

```python
df.drop("Duration",axis=1,inplace=True)
```

```python
df.head(1)
```

```python
arr=encoder.fit_transform(df[["Airline","Source","Destination"]]).toarray();
df1=pd.DataFrame(arr,columns=encoder.get_feature_names_out());
```

```python
df.drop("Airline",axis=1,inplace=True)
df.drop("Source",axis=1,inplace=True)
df.drop("Destination",axis=1,inplace=True)
```
