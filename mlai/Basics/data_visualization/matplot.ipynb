```python
import pandas as pd;
import matplotlib.pyplot as plt;
```

```python
df=pd.read_csv("customer_shopping_data.csv")
df.info()
```

```python
df["payment_method"].unique()
```

```python
x=df[df["payment_method"]=="Credit Card"]
y=df[df["payment_method"]=="Debit Card"]
z=df[df["payment_method"]=="Cash"]
```

```python
plt.figure(figsize=(9,5))
plt.subplot(1,3,1)
plt.plot()
plt.subplot(1,3,2)
plt.plot()
plt.subplot(1,3,3)
plt.plot()
```

```python
plt.hist([x["price"], y["price"], z["price"]], bins=1, label=["Credit Card", "Debit Card", "Cash"], color=["blue", "green", "orange"], alpha=0.7)
plt.title("Price Distribution by Payment Method")
plt.xlabel("Price")
plt.ylabel("Frequency")
# plt.legend()
plt.show()
```
