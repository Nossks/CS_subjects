```python
from sklearn.datasets import make_classification
```

```python
X,Y=make_classification(n_samples=500,n_redundant=0,n_features=2,n_clusters_per_class=1,
weights=[.9],random_state=13)
```

```python
import pandas as pd
```

```python
ip=pd.DataFrame(X,columns=["c1","c2"])
op=pd.DataFrame(Y,columns=["result"]);
```

```python
df=pd.concat([ip,op],axis=1)
```

```python
df["result"].value_counts()
```

```python
import matplotlib.pyplot as mlt;
mlt.scatter(df["c1"],df["c2"],c=df["result"])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
```

```python
newX,newY=oversample.fit_resample(df[["c1","c2"]],df["result"])
resampled_df=pd.concat([newX,newY])
```

```python
resampled_df=pd.concat([newX,newY],axis=1)
```

```python
resampled_df["result"].value_counts()
```

```python
mlt.scatter(resampled_df["c1"],resampled_df["c2"],c=resampled_df["result"])
```
