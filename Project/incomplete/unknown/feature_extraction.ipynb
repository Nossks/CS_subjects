```python
import os
import cv2 as cv
from PIL import Image
from img2vec_pytorch import Img2Vec
import matplotlib.pyplot as plt
import pandas as pd

img2vec=Img2Vec(cuda=False)
```

```python
data="/home/cry-more/cv/pro/extraction_data/"
```

```python
os.listdir(data)
```

```python
categories=["real","fake"]
```

```python
X=[]
y=[]
for j,category in enumerate(categories):
    for file in os.listdir(os.path.join(data,category)):
        img=Image.open(os.path.join(data,category,file))
        img_feature=img2vec.get_vec(img)

        X.append(img_feature)
        y.append(j)
```

```python
pd.DataFrame(X).shape,pd.DataFrame(y).shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=3)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rr=RandomForestClassifier()

rr.fit(X_train,y_train)
```

```python
y_pred=rr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
accuracy_score(y_pred,y_test)
```
